```python
def addition(values):
    '''This function returns the sum of the list variables'''
    sum=0
    for value in values:
        sum+=value
    return sum
l=[1,2,3,4,5]
print(addition(l))
```

```python
def divisibility(n):
    '''This function returns the values which are divisible by 2,3,5'''
    for i in range(2,n+1):
        if(i%2==0 or i%3==0 or i%5==0):
            print(i,end=',')
divisibility(10)
```

```python
def countuclc(txt):
    countuc=0
    countlc=0
    for t in txt:
        if ord(t)>=65 and ord(t)<=90:
            countuc+=1
        elif ord(t)>=97 and ord(t)<=122:
            countlc+=1
    print("No. of Upper case char: ",countuc)
    print("No. of Lower case char: ",countlc)
countuclc("JigSaw Academy")
```

```python
def validate_password(password):
    '''This function validates passwords i.e it checks '''
    tries_left=3
    while(tries_left>0):
        uppercase_count=lowercase_count=spl_char_count=number_count=0
        for p in password:
            if p.isupper():
                uppercase_count+=1
            elif p.islower():
                lowercase_count+=1
            elif p.isdigit():
                number_count+=1
            elif p=='!' or p=='@' or p=='#' or p=='$' or p=='%' or p=='^' or p=='&' or p=='*' or p=='?' or p=='_':
                spl_char_count+=1
        if uppercase_count and lowercase_count and spl_char_count and number_count:
            print("Password Validated")
            break
        else:
            print("RETRY.",tries_left-1,"attempts left")
        tries_left-=1
password=input("Enter your password")
validate_password(password)
```

```python
def validate_password():
    '''This function validates passwords'''
    tries_left=3
    while(tries_left>0):
        password=input("Enter your password")
        uppercase_count=lowercase_count=spl_char_count=number_count=0
        length=len(password)
        for p in password:
            if p.isupper():
                uppercase_count+=1
            elif p.islower():
                lowercase_count+=1
            elif p.isdigit():
                number_count+=1
            elif p in "!@#$%^&*<>?/\|{}[]_":
                spl_char_count+=1
        if uppercase_count and lowercase_count and spl_char_count and number_count and length<=8:
            print("Password is Valid")
            break
        else:
            print("RETRY.",tries_left-1,"attempts left")
        tries_left-=1
validate_password()
```

```python
def palindrome(txt):
    txt_palindrome=''
    for c in txt:
        txt_palindrome=c+txt_palindrome
    if(txt==txt_palindrome):
        print("palindrome")
    else:
        print("Not a palindrome")
txt=input("Enter a word: ")
palindrome(txt)
```

```python
class list_operations():
    def __init__(self,li):
        self.li=li
    def append(self,value):
        return self.li+[value]
        #return self.li
    def pop(self):
        return self.li[:-1]
        #return self.li
    def count(self,value):
        count_item=0
        for element in self.li:
            if element==value:
                count_item+=1
        return count_item
    def reverse(self):
        return self.li[::-1]
        #return self.li
    def extend(self,l):
        return self.li+l
        #return self.li
obj=list_operations([1,2,3,4,5])
print(obj.append(6))
print(obj.pop())
print(obj.count(1))
print(obj.reverse())
print(obj.extend([6,7]))
```
