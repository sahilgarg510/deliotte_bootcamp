```python
def calculator(n,m,x):
    if x=='+':
        print(n+m)
    elif x=='-':
        print(n-m)
    elif x=='*':
        print(n*m)
    elif x=='/':
        print(n/m)
    else:
        print('error')
calculator(2,2,'+')
```

```python
#create a function which can find the sum of list variable
#list =[1,2,3,4,5]= 15
def addition(values):
    sum=0
    for i in values:
        sum=sum+i
    return sum
values=[1,2,3,4,5]
addition(values)
```

```python
#create a function which can return us values which are divisible by 2,3,5
def divisibility(n): 
    for i in range(1,n+1):
        if(i%2==0 or i%3==0 or i%5==0):
            print(i)
n=int(input())
divisibility(n)
```

```python
def upper_lower(txt):
    uppercase=0
    lowercase=0
    for i in txt:
        if i.isupper():
            uppercase=uppercase+1
        elif i.islower():
            lowercase=lowercase+1
    print(uppercase)
    print(lowercase)
txt="Create a function which can return the number of uppercase and lowercase letters in text"
upper_lower(txt)
```

```python
#create a function which can print the upper matrix values as table of two
k=1
for i in range(0,3):
    for j in range(0,3):
        if(i>j):
            print(" ",end=" ")
        else:
            print(2*k, end=' ')
            k+=1
    print("\n")
```

```python
#create a function whcih can validate the passwords
#uppercase, lowercase, special character, digits
#len-8
#person have 3 trials to verify
def validate_password(password):
    uppercase=0
    lowercase=0
    specialchar=0
    if len(password)>=8:
        for i in password:
            if i.isupper():
                uppercase=uppercase+1
            if i.islower():
                lowercase=lowercase+1
            if i=='@' or i=='#' or i=='!' or i=='$' or i=='%' or i=='^' or i=='&' or i=='*':
                specialchar=specialchar+1
    if uppercase> 0 and lowercase> 0 and specialchar> 0: 
        print("Valid")
    else:
        print("Invalid")
password="Abcs123#"
validate_password(password)
```

```python
def palindrome(name):
    reversed_name=name[::-1]
    if (name==reversed_name):
        print("Is a Palindrome")
    else:
        print("Is not Palindrome")
name="malayalam"
palindrome(name)
```

```python
class ListValue:
    
    def __init__(self, values):
        self.values=values
        
    def append1(self, newvalue):
        self.values.append(newvalue)
    
    def extend1(self, newvalue1):
        self.values.extend(newvalue1)
        
    def index1(self, newvalue2):
        return  self.values.index(newvalue2)
    def pop1(self, newvalue3):
        self.values.pop(newvalue3)
        
```
