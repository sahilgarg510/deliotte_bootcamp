```python
def password_validation(string):
    lowcount=0
    upcount=0
    specount=0
    numcount=0
    if len(string)>8:
        for i in string:
            if i.isupper():
                upcount+=1
            if i.islower():
                lowcount+=1
            if i.isdigit():
                numcount+=1
            if i=="#" or i=="@"  or i=="$" or i=="!" or i=="^" or i=="&" :
                specount+=1
            if upcount>=1 and lowcount>=1 and numcount >=1 and specount >= 1 :
                res= 1
            else:
                res=0
    else:
        res=0
    if res==1:
        print("Strong password")
    else:
        print("Invalid Password")
        
password_validation("$Gentleman1")

        
```

```python
def palin(txt):
    rtxt=txt[::-1]
    if txt==rtxt:
        print("String is a Palindrome")
    else:
        print("String is not a Palindrome")
palin("madam")
```

```python
class listmethods:
    def __init__(self,elements):
        self.elements=elements
    def liscopy(self):
        cplist=self.elements.copy()
        return cplist
    def lisreverse(self):
        rlist=self.elements[::-1]
        return rlist
    def lisappend(self,newelement):
        newlist=self.elements.append(newelement)
        return newlist
    def lispop(self,indexno):
        slist=self.elements.pop(indexno)
        return slist
    def lisextend(self,val):
        return self.elements.extend(val)
    def lisindex(self,indexno):
         return self.elements.index(indexno)
    
```

```python
lis1=listmethods([1,2,3,4])
```

```python
listmethods.liscopy(lis1)
```
