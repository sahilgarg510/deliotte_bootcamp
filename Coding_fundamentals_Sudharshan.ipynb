```python
def pwd_validated(txt):
    times=3
    while(times>0):
        if(times<=2):
            txt=input("Enter password again")
        countspl=0
        countup=0
        countdigit=0
        countlow=0
        total=0
        length_pwd=len(txt)
        for i in range(0,length_pwd):
            if(txt[i].isupper()):
                countup=countup+1
            elif(txt[i].islower()):
                countlow=countlow+1
            elif(txt[i].isdigit()):
                countdigit=countdigit+1
            elif(txt[i]=="!" or txt[i]=="@" or txt[i]=="#" or txt[i]=="$" or txt[i]=="%" or txt[i]=="^" or txt[i]=="&" ):
                countspl=countspl+1
        if(countup>=1 and countlow>=1 and countdigit>=1 and countspl>=1 and length_pwd<=8):
            print("The password is correct")
            break
        else:
            times=times-1
            print(f"Not Correct, {times} attempts left")
password=input("Enter the password:")
pwd_validated(password)
     
```

```python
class Listt:
    def __init__(self,lis):
        self.lis=lis
    def appendd(self,val):
        self.lis=self.lis+[val]
        return self.lis
    def countt(self):
        count=0
        for i in self.lis:
            count=count+1
        return count
    def indexx(self,val):
        count=-1
        for j in self.lis:
            count=count+1
            if(j==val):
                break
        return count
    def reversee(self):
        self.lis=self.lis[::-1]
        return self.lis
    def copyy(self):
        liscopy=[]
        for k in self.lis:
            liscopy=liscopy+[k]
        return liscopy
a=Listt([1,2,3])
print(a.appendd(4))
print(a.countt())
print(a.indexx(2))
print(a.reversee())
print(a.copyy())

```
