```python
class List2:
    
    def __init__(self, values):
        self.values=values
        
    def add(self, newvalue):
        self.values.append(newvalue)
        
    def count1(self):
        self.values.count()
        
    def move(self,index):
        self.values.pop(index)
```

```python
li1 = List2([1,2,3,4])
```

```python
li1.add(5)
```
