```python
class listreplication:
    def __init__ (self,values):
        self.values=values
        
    def append(self,value1):
        self.values.append(value1)
        
    def insert(self,newnum):
        self.values.insert(newnum)
        
    def reverse(self,value2):
        self.values.reverse(value2)
        
    def remove(self,value4):
        self.values.remove(value4)
    
    def extend(self, value5):
        self.values.extend(value5)
```

```python
#to check whether the number is divisible by 2,3 or 5

def divisibility(number):
    for i in range(number):
        if(i%2==0 or i%3==0 or i%5==0):
            print(i)
        
divisibility(10)
```

```python
#function to count number of upper and lowercase aplabets in a word

def numofchar(text):
    upper=0
    lower=0
    for i in text:
        if i.isupper():
            upper=upper+1
        elif i.islower():
            lower=lower+1
            
    print('uppercase:',upper)
    print('lowercase:',lower)
        
numofchar('mAdhu')
```

```python
# to check whether the given name is palindrome or not

def name_palindrome(name):

#reversing the name and comparing with the original     
    
    if(name==name[::-1]):
        print('the name is palindrome')
    else:
        print(' not palindrome')

#calling function

name_palindrome('madhu')
```

```python
#function to check whether the passcode is valid or not

def validate_password(passcode):
    lower=0
    upper=0
    digit=0
    schar=0
    splchar='!@#$%^&*'
    
#checking if length of password is atleast 8 or not
    
    if (len(passcode)>=8):
        for i in passcode:
            if(i.islower()):
                lower=lower+1
            if(i.isupper()):
                upper=upper+1
            if(i.isdigit()):
                digit=digit+1
            if(i in splchar):
                schar=schar+1
    if(lower>=1 and upper>=1 and digit>=1 and schar>=1 and lower+upper+digit+schar==len(passcode)):
        print("valid password")
    else:
        print("invalid password")
        
#calling function        
validate_password('Welcome@123')
                
        
```
