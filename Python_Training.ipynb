```python
# function to return number of lower cases and upper cases in text

def Upper_lower_count(text):
    lower_case = 0
    upper_case = 0
    for i in text:
        if i.islower():
            lower_case += 1
        elif i.isupper():
            upper_case += 1
    return lower_case,upper_case

Upper_lower_count("Boot Camp")
        
```

```python
# create a class object which replicates the list class

class ListClass:
    def __init__(self, values):
        self.values = values
    
    def append(self,item2):
        self.values.append(item2)
        print(self.values)
    
    def index(self,v):
        print(self.values.index(v))
    
    def remove(self,v):
        self.values.remove(v)
        print(self.values)
    
    def count(self,v):
        print(self.values.count(v))
        
l = [1,"Python",1,'Book',1]
List1 = ListClass(l)

List1.append("it")

List1.index('Python')

List1.remove("Book")

List1.count(1)
```
