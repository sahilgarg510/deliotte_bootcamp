```python
#Calculator using if-elif-else

a = int(input("Enter the 1st value: "))
b = int(input("Enter the 2nd value: "))
operation = input("Enter operator: ")
#Addition
if operation == "+":
    print(a+b)
#Subtraction
elif operation == "-":
    print(a-b)
#Multiplication
elif operation == "*":
    print(a*b)
#Division
elif operation == "/":
    print(a/b)
else:
    print("Invalid Operation")
```

```python
#Sum of a list variable
def sum_list(lst):
    sum = 0
    for num in lst:
        sum += num
    return sum
lst = [1,2,3,4,5]
sum_list(lst)
```

```python
#Numbers divisble by 2,3,5 between numbers in a range
def divisibility(n):
    for i in range (n+1):
        if (i%2==0 or i%3==0 or i%5==0):
            print(i)
divisibility(10)
```

```python
#Function to return number of upper cases and lower cases
def text_case(word):
    upper_case = 0
    lower_case = 0
    for i in word:
        if i == i.upper():
            upper_case += 1
        if i == i.lower():
            lower_case += 1
    print(f"Number of Upper cases: {upper_case}\nNumber of Lower Cases:{lower_case}")
    
text_case("EXpeCTaTionS")
    
    
```

```python
#Function which can print the upper matrix with multiples of two
def upper_matrix(n):
    k = 1
    for i in range(4):
        for j in range (4):
            if i>j:
                print(" ",end = " ")
            else:
                print(n*k,end = " ")
                k = k+1
        print("\n")
            
                
        
upper_matrix(2)            
```

```python
#Validate Password

def validate_password(password):
    dig_char_list = ['@','#','$','%','^','&','*'] 
    for i in password:
        if (i.isupper()) or (i.islower()) or (i.isdigit()) or (i in dig_char_list):
            if len(password)>=8:
                return('Valid Password')
            else:
                return('Invalid Password')
        
        
        
validate_password('M22&@_')
                
```

```python
#Palindrome

def palindrome(text):
    if text == text[::-1]:
        print('Its a palindrome')
    else:
        print('Its not a palindrome')
        
palindrome('SNSM')
        
```

```python
#List Methods

class List_Methods:
    
    def __init__(self,lst):
        self.lst = lst
    
    def add(self,value):
        return self.lst.append(value)
        
    def _remove(self):
        value = input('Enter the item you want to remove: ')
        self.lst = self.lst.remove(value)
        return self.lst
    def _count(self):
        value = input('Enter the item you want to count: ')
        return self.lst.count(value)
    def _reverse(self):
        return self.lst.reverse()
    def _clear(self):
        return self.lst.clear()
```

```python
#List Methods

class ListMethods:
    
    def __init__(self,lst):
        self.lst = lst
    
    def add(self,value):
        lst = self.lst.append(value)
        return self.lst
        
    def _remove(self,value):
        lst = self.lst.remove(value)
        return self.lst
    
    def _count(self,value):
        return self.lst.count(value)
    
    def _reverse(self):
        return self.lst.reverse()
    
    def _clear(self):
        return self.lst.clear()
```

```python
lst1 = ListMethods([1,2,3,4,5])
print(lst1.add(3))
print(lst1._remove(2))
print(lst1._count(3))
print(lst1._reverse())
print(lst1._clear())
```
